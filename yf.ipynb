```python
paths = glob.glob('./y*/*.dat', recursive=True)
```

```python
def get_folder_name(path):
    return path.split('/')[1]
```

```python
for path in paths:
    df = pd.read_csv(path, header=None, sep=' ', usecols=[6, 12])
    x = df.iloc[:, 0].values
    y = df.iloc[:, 1].values
    fig, ax = plt.subplots()
    (markerline, stemlines, baseline) = ax.stem(x, y, markerfmt=' ', basefmt=' ')
    ax.set_xlim(0,30)
    ax.set_ylim(0,11000)
    fig.savefig('{}.png'.format(get_folder_name(path)))
```
